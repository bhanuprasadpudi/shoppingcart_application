```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        # return '{:s}, {:.2f}'.format(self.name, self.price)
        return f'{self.name}, {self.price:.2f}'

class ShoppingCart:
    def __init__(self):
        self.items = []
        self.wishlist = []

    def add(self, item):
        self.items.append(item)

    def update(self, name, price):
        for item in self.items:
            if item.name == name:
                item.price = price

    def wishlist_add(self, item):
        self.wishlist.append(item)
    
    def wishlistitems(self):
        for v in self.wishlist:
            print(v)
            
    def cart_wishlist(self, name):
        for item in self.wishlist:
            if item.name == name:
                self.items.append(item)
        
    def total(self):
        return sum(item.price for item in self.items)

    def whishlist_total(self):
        return sum(item.price for item in self.wishlist)

    def remove_items(self, name):
        self.items = [item for item in self.items if item.name != name]

    def remove_whislist_items(self, name):
        self.wishlist = [item for item in self.wishlist if item.name != name]
        
    def __str__(self):
        return '\n'.join(str(item) for item in self.items)

user_choice='''
Your choice:
a - add item
l - list item
u - update item
w - add item to wishlist
wl- wishlist items
cw - add item to cart from whishlist
wr - remove item form whislist
t - total price of items in cart
wt - total price of wishlist items
r - remove an item
q - quit
enter here :
'''
    
cart = ShoppingCart()
def menu():
    user_input = input(user_choice)
    while user_input != 'q':
        if user_input == 'a':
            cart.add(item = Item(input('item name:'), float(input('item price:'))))
        elif user_input == 'l':
            print(cart)
        elif user_input == 'r':
            cart.remove_items(input('item name:'))
        elif user_input == 't':
            print(cart.total())
        elif user_input == 'u':
            cart.update(input('item name:'), float(input('new price:')))
        elif user_input == 'c':
            cart.clear_items()
        elif user_input == 'w':
            cart.wishlist_add(item = Item(input('item name:'), float(input('item price:'))))
        elif user_input == 'wl':
            cart.wishlistitems()
        elif user_input == 'cw':
            cart.cart_wishlist(input('item name:'))
        elif user_input == 'wr':
            cart.remove_whislist_items(input('item name:'))
        elif user_input == 'wt':
            print(cart.whishlist_total())
        else:
            print('Unknown command, Please try again!')
        user_input = input(user_choice)

menu()
```
